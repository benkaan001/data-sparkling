## Step 1. Create DF for GDP series for all countries
  - Identify duplicate series
  - Filter df using `map` and `lambda`

```python
fred.search('country GDP')
```

```python
# create the df and query df titles
gdp_df = fred.search('country GDP').query('frequency == "Annual" ')
gdp_df['title']
```

- GDP data title for a country appears to start with `Gross Domestic Product for <CountryName>`

```python
# explore the titles 

for index,title in enumerate(gdp_df['title']):
    if title.startswith('Gross Domestic Product for '):
        print(index, title)
```

### Step 2. Filter the DF to include only ID's starting with `MKTGDP...`

```python
criterion2 = gdp_df['id'].map(lambda x: x.startswith("MKTGDP"))
gdp_df[criterion2]
```

```python
# check the titles
for title in gdp_df[criterion2]['title']:
    print(title)
```

```python
# rewrite the df
gdp_df = gdp_df[criterion2]
```

### Step 3. Merge the GDP Data for Countries 
  - Convert country-specific series into df 
  - Join country-specific dfs into one
  - Extract the country names from df `title`

```python
all_countries = []

try:
    for id in gdp_df.index:
        country_series = fred.get_series(series_id=id)
        country_df = country_series.to_frame(name=id)
        all_countries.append(country_df)
except Exception as e:
    print(e)
else:
    print('Data is ready!')
```

```python
all_countries
```

```python
# check the size of the list
len(all_countries)
```

```python
# concatenate the data based on column value
all_countries_df = pd.concat(all_countries, axis=1)
all_countries_df
```

- Update the `id` value with `country name`

```python
# update the id with country name
id = 'MKTGDPCNA646NWDB'
title = gdp_df['title'][id]
title 
```

```python
title = title.replace("Gross Domestic Product for ", "")
title
```

```python
#  function to be passed into the df.rename()
def rename_column(col_id):
    country_name = gdp_df['title'][col_id].replace("Gross Domestic Product for ","")
    return country_name

# test run
rename_column(id)
```

```python
# rename the df columns 
all_countries_df = all_countries_df.rename(rename_column, axis=1)
all_countries_df
```

```python
# drop World column
all_countries_df = all_countries_df.drop(columns='World (DISCONTINUED)', axis=1)
```

```python
all_countries_df.shape
```

### Step 4. Finalize DF
  - Find and fill missing values with 0
  - Typecast the rows to INT
  - Display the final results in billions

```python
# find total count of missing values
all_countries_df.isna().sum()
```

```python
# replace missing values with 0
all_countries_df = all_countries_df.fillna(0)
```

```python
# convert the data to represent it in $billions
all_countries_df = all_countries_df.astype(int)/1_000_000_000
all_countries_df
```

```python
# create a 'column' date consisting of index values
all_countries_df.insert(loc=0, column='date', value=all_countries_df.index)
all_countries_df
```

```python
# deep copy and rename the df
world_gdp_df = all_countries_df.copy()
world_gdp_df
```
