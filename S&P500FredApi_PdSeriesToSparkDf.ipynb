## Step 2. Create Spark DF from Pandas DF
- Type caste index_date column to `date`
- Find out about missing values

```python
# create spark DF
sp500_df = spark.createDataFrame(sp500_pdf)

display(sp500_df)
```

```python
# get schema info
sp500_df.printSchema()
```

```python
# convert the index_date column from timestamp to date type 
sp500_df = sp500_df.withColumn("index_date", sp500_df.index_date.cast("date"))

sp500_df.printSchema()
```

```python
# display the 5 missing 'index_value' column values
sp500_df.filter(sp500_df.index_value.isNull()).show(5)
```

```python
# display the count of missing values for index_date column
sp500_df.filter(sp500_df.index_date.isNull()).count()
```

## Step 3. Query Spark DF Using Spark SQL
- Create a table
- Get statistics 
- Run SQL queries against the DF

```python
# create a table using sp500_df
sp500_df.write.mode("overwrite").option("overwriteSchema", "true").saveAsTable("sp500")

# get metadata info
spark.sql("DESCRIBE EXTENDED sp500").show()
```

```python
# run sample queries
q1 = """ 
            SELECT ROUND(AVG(index_value), 2) index_average
            FROM sp500 
            
            """

q2 = """ 
             SELECT MAX(index_value) highest_performance, 
                    MIN(index_value) lowest_performance
             FROM sp500
             
             """

q3 = """ 
            SELECT COUNT(*) null_count
            FROM sp500
            WHERE index_value IS NULL
            
            """

q4 = """
    SELECT index_date, 
        CASE 
            WHEN index_value == (SELECT MAX(index_value) FROM sp500) THEN 'best day ever'
            WHEN index_value == (SELECT MIN(index_value) FROM sp500) THEN 'wort day ever'
            WHEN index_value > (SELECT ROUND(AVG(index_value),2) FROM sp500) THEN 'above average'
            WHEN index_value < (SELECT ROUND(AVG(index_value),2) FROM sp500) THEN 'below_average'
            ELSE 'just another day'
        END AS daily_commentary
    FROM sp500

"""

queries = [q1, q2, q3, q4]

for query in queries:
    spark.sql(query).show()
```
